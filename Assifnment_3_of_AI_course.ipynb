## Question no.4:
Design a Python program that compares two text inputs and determines their
similarity using the OpenAI API.

```python
# Initializing the genini model
model = genai.GenerativeModel(model_name="gemini-1.5-flash-latest",
                              generation_config=generation_config,
                              system_instruction="You will check the similarity between two text prompts and give a similarity score based on that. You will give only similarity score in a range of 0 to 1(2 decimal points). No other details about the text\n",
                              safety_settings=safety_settings)

# Initializing the conversation
convo = model.start_chat(history=[
])

# Getting the input from the user
text1 = str(input("Enter your text 1 here: "))
text2 = str(input("Enter your text 2 here: "))

# Creating a function that uses gemini model to check the similarity between the given texts and also handling the exceptions
def similarity_analyzer(text1, text2):
    try:
       if not text1 or not text2:
          raise ValueError("Both text inputs must be provided.")
       convo.send_message(text1 + ',' + text2)
       print(convo.last.text)
       print('Similarity score ranges form 0 to 1(0 means that the given texts are totaly unsimilar while 1 means that the given texts are totally similar)')
    except Exception as e:
       print(f"An unexpected error occurred: {e}")

# Calling the function
similarity_analyzer(text1, text2)
```

## Question no.9:
Create a text generator that produces creative text continuations based on your input
prompts.

```python
# Initialize the text generation pipeline with a pre-trained model
generator = pipeline('text-generation')
conversation = []
```

```python
# Function to generate text based on input prompt
def generate_text(prompt):
    # Generate text continuations
    generated_texts = generator(prompt, max_length=500, num_return_sequences=1)
    return generated_texts[0]['generated_text']

# Example usage
prompt = input("Enter your prompt here: ")
conversation.append({"role": "user", "content": prompt})
print(conversation)
generated_text = generate_text(conversation)
conversation.append({"role": "assistant", "content": generated_text})

print("Generated Text: ", generated_text)
print(conversation)
```

## Question no.10:
Use any one hugging face model in your project. - Write a python code that uses
HuggingFace tokenizers library to tokenize a given sentence using the BERT tokenizers.


```python
# Initialize the BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
# Input sentence
sentence = input("Enter a sentence to tokenize: ")
# Tokenize the sentence
tokens = tokenizer.tokenize(sentence)
# Display the tokens
print("Tokens:", tokens)
# Convert tokens to token IDs
token_ids = tokenizer.convert_tokens_to_ids(tokens)
# Display the token IDs
print("Token IDs:", token_ids)
```
